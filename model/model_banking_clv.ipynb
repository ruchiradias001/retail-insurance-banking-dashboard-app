```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
df=pd.read_excel("Prediction Data.xlsx")
df.head()
```

```python
df.rename(columns={'Customer Lifetime Value':'CLV'},inplace=True)
```

```python
#label encoding 
label_encoder = preprocessing.LabelEncoder()
df['Credit card holder']= label_encoder.fit_transform(df['Credit card holder']) 
```

```python
df_new= df.drop(['Customer ID', 'Marital status','Branch ID','Branch Name','Gender'], axis = 1)
df_new.head()
```

```python
X=df_new.drop(['CLV'],axis=1)
y=df_new['CLV']          
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
#Fitting random forest model
model = RandomForestRegressor(random_state=1)
model.fit(X_train,y_train)
```

```python
#Saving model to pickle file
pickle.dump(model, open("model_banking_clv.pkl", "wb"))
```

```python
pred = model.predict([[7700, 34, 6500, 8000, 2, 1, 1, 1, 2000]])
print(pred[0])
```
